```python
import pandas as pd
import numpy as np
```

```python
url = ("https://raw.github.com/pandas-dev/pandas/main/pandas/tests/io/data/csv/tips.csv")
tips = pd.read_csv(url)
tips
```

```python
'''SELECT'''

# SELECT total_bill, tip, smoker, time
# FROM tips;

tips[["total_bill", "tip", "smoker", "time"]]
```

```python
'''SELECT & NEW COLUMN'''

# SELECT *, tip/total_bill as tip_rate
# FROM tips;

tips.assign(tip_rate=tips.tip/tips.total_bill)
```

```python
'''WHERE(Filtering)'''

# SELECT *
# FROM tips
# WHERE time = 'Dinner';

tips[tips.total_bill>10]
```

```python
is_dinner = tips["time"] == "Dinner"
is_dinner.value_counts()
```

```python
tips.loc[tips["time"] == "Dinner"]
tips[is_dinner] # Return TRUE index
```

```python
'''NULL VALUE'''

# SELECT *
# FROM frame
# WHERE col2 IS NULL;

tips[tips.tip.isna()]
tips[tips.tip.notna()]
```

```python
'''GROUP BY'''
# SELECT sex, count(*)
# FROM tips
# GROUP BY sex;

tips.groupby("sex").size()
```

```python
# Notice that in the pandas code we used size() and not count(). 
# This is because count() applies the function to each column, returning the number of NOT NULL records within each.

# tips.groupby("sex")["total_bill"].count()
tips.groupby("sex").count()
```

```python
tips
```

```python
# SELECT day, AVG(tip), COUNT(*)
# FROM tips
# GROUP BY day;

tips.groupby("day").agg({"tip": np.mean, "day": np.size})
```

```python
# SELECT smoker, day, COUNT(*), AVG(tip)
# FROM tips
# GROUP BY smoker, day;

tips.groupby(["smoker", "day"]).agg({"tip": [np.size, np.mean]})
```

```python
'''Join'''

df1 = pd.DataFrame({"key": ["A", "B", "C", "D"], "value": np.random.randn(4)})
df2 = pd.DataFrame({"key": ["B", "D", "D", "E"], "value": np.random.randn(4)})
```

```python
df1
```

```python
df2
```

```python
# SELECT *
# FROM df1
# INNER JOIN df2
#   ON df1.key = df2.key;

pd.merge(df1, df2, on="key") # inner join
```

```python
indexed_df2 = df2.set_index("key")
pd.merge(df1, indexed_df2, left_on="key", right_index=True)
```

```python
# SELECT *
# FROM df1
# LEFT OUTER JOIN df2
#   ON df1.key = df2.key;

pd.merge(df1, df2, on="key", how="left")
```

```python
# SELECT *
# FROM df1
# RIGHT OUTER JOIN df2
#   ON df1.key = df2.key;

pd.merge(df1, df2, on="key", how="right")
```

```python
# SELECT *
# FROM df1
# FULL OUTER JOIN df2
#   ON df1.key = df2.key;

pd.merge(df1, df2, on="key", how="outer")
```

```python
df1 = pd.DataFrame({"city": ["Chicago", "San Francisco", "New York City"], "rank": range(1, 4)})
df2 = pd.DataFrame({"city": ["Chicago", "Boston", "Los Angeles"], "rank": [1, 4, 5]})
```

```python
'''Union'''

# SELECT city, rank
# FROM df1
# UNION ALL
# SELECT city, rank
# FROM df2;

pd.concat([df1, df2]) # union all
pd.concat([df1, df2]).drop_duplicates() # union
```
